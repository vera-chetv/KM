##### Получение функций перемещения из второго закона Ньютона

Рассмотрим, как из второго закона Ньютона и начальных условий можно получить аналитические выражения для функций перемещения $s_x(t)$ и $s_y(t)$ для движущегося тела. Начнем с определения системы координат и условий задачи.

***Исходные условия:***

Предположим, что начало прямоугольной декартовой системы координат соответствует уровню земли. В начальный момент времени $t = 0$ задаются следующие условия:

- $s_x(0) = 0$ (начальное положение по оси X)
- $s_y(0) = h_{\text{start}}$ (начальное положение по оси Y)

Тело движется под воздействием силы тяжести, поэтому мы рассмотрим его движение в вертикальном и горизонтальном направлениях.

***Второй закон Ньютона:***

Согласно второму закону Ньютона, сила, действующая на тело, связана с его массой и ускорением:

$F = ma$, где:
- $F$ — приложенная сила,
- $m$ — масса тела,
- $a$ — ускорение тела.

Для случая свободного падения, когда тело движется только под воздействием силы тяжести, мы можем записать:

$F = m(0, -g)$, где $g$ — ускорение свободного падения (приблизительно $9.81 \, \text{м/с}^2$).

***Уравнения движения***

Из второго закона Ньютона мы можем записать уравнения движения по осям X и Y:

$ma$ = $s''_x(t)$, $s''_y(t)$, где:
- $s''_x(t) = 0$ (ускорение по оси X, поскольку нет горизонтальных сил),
- $s''_y(t) = -g$ (ускорение по оси Y, равное ускорению свободного падения).

Таким образом, мы получаем два уравнения:

$s''_x(t) = 0$

$s''_y(t) = -g$

**Решение уравнений движения**

***Уравнение по оси X***

Для оси X у нас есть:

$s''_x(t) = 0$

Интегрируя это уравнение дважды, получаем:

1. Первая интеграция:

$s'_x(t) = v_{0x} \quad + C_1$, где $v_{0x}$ — начальная скорость по оси X, а $C_1$ — константа интегрирования.

2. Вторая интеграция:

$s_x(t) = v_{0x} t + C_2$, где $C_2$  — еще одна константа интегрирования, которая равна $s_x(0) = 0$

$s_x(t) = v_{0x} t$

***Уравнение по оси Y***

Для оси Y у нас есть:

$s''_y(t) = -g$

Интегрируя это уравнение дважды, получаем:

1. Первая интеграция:

$s'_y(t) = -gt + v_{0y} + C_3$, где $v_{0y}$ — начальная скорость по оси Y, а $C_3$ — константа интегрирования.

2. Вторая интеграция:

$s_y(t) = -\frac{g}{2} t^2 + v_{0y} t + C_4$, где $C_4$ равна $s_y(0) = h_{\text{start}}$

$s_y(t) = h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2$

**Итоговые функции перемещения**

Таким образом, мы получили выражения для функций перемещения:

$s_x(t) = v_{0x} t$

$s_y(t) = h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2$,

где $v_0 = (v_{0x}, v_{0y})$ обозначает начальную скорость движения тела в момент времени $t = 0$.

**Изменение функций перемещения**

##### Этап 1. Определение функции перемещения по оси Y

Запишем функцию перемещения по оси Y:

$s_y(t) = h_{\text{start}} + v_{0y} t - \frac{g}{2} t^2$, где
- $h_{\text{start}}$ — начальная высота,
- $v_{0y}$ — начальная скорость по оси Y,
- $g$ — ускорение свободного падения.

##### Этап 2. Определение начальной скорости по оси Y

Конечная высота в момент времени $T$ определяется как:

$$h_{\text{end}} = h_{\text{start}} + v_{0y} T - \frac{g}{2} T^2$$

##### Этап 3. Решение для $v_{0y}$

Подставляем известные значения:

$$h_{\text{end}} = h_{\text{start}} + v_{0y} T - \frac{g}{2} T^2$$

##### Этап 4. Построение последовательности $s_y$
##### Этап 5. Уравнение прямой $S_{\text{lin}}(t)$

Прямая, проходящая через точки $(t[k], s_y[k])$ и $(t[k + 1], s_y[k + 1])$:

$$S_{\text{lin}}(t) = s_y[k] + \frac{s_y[k + 1] - s_y[k]}{t[k + 1] - t[k]} (t - t[k])$$

##### Этап 6. Поиск значения $t^*$

Для нахождения $t^*$ из условия $S_{\text{lin}}(t^*) = 0$:

$$t^* = t[k] - \frac{s_y[k] (t[k + 1] - t[k])}{s_y[k + 1] - s_y[k]}$$

##### Этап 7. Точное решение

Уравнение для точного значения $t^*$:

$$s_y(t^*) = 0 \implies h_{\text{start}} + v_{0y} t^* - \frac{g}{2} (t^*)^2 = 0$$

##### Этап 8. Относительная ошибка

Относительная ошибка между точным и приближенным решениями:

$$\text{Относительная ошибка} = \left| \frac{t^*_{\text{точное}} - t^*_{\text{приближенное}}}{t^*_{\text{точное}}} \right| \times 100\%$$

```python
t_values = np.linspace(0, T, 100)
s_x_values = (s_end / T) * t_values  # Горизонтальное перемещение
s_y_values = h_start + v0_y * t_values - 0.5 * g * t_values**2  # Вертикальное перемещение

# Поиск индекса k
k = None
for i in range(len(s_y_values) - 1):
    if s_y_values[i] * s_y_values[i + 1] < 0:
        k = i
        break

# Проверка, найден ли индекс k
if k is not None:
    t_k = t_values[k]
    s_y_k = s_y_values[k]
    t_k_plus_1 = t_values[k + 1]
    s_y_k_plus_1 = s_y_values[k + 1]
    
    print(f"Пересечение с осью Y на t = {t_k}, s_y = {s_y_k}")
else:
    print("Нет пересечения с осью Y.")
```

#### Угловой коэффициент $k$
Угловой коэффициент $k$ для касательной прямой определяется как производная функции в точке, где мы хотим построить касательную. В данном случае:

1. **Скорость по оси X**: Это общая горизонтальная скорость, вычисленная как $\frac{s_{end}}{T}$.
2. **Скорость по оси Y**: Это вертикальная скорость в момент времени $t = 0$, равная начальной скорости $v_{0y}$.

Угловой коэффициент $k$ вычисляется по формуле:

$$k = \frac{s_{y\_derivative}}{s_{x\_derivative}},$$ где:
- $s_{y\_derivative}$ — производная по оси Y (вертикальная скорость),
- $s_{x\_derivative}$ — производная по оси X (горизонтальная скорость).

#### Уравнение нормальной прямой к траектории движения тела в момент времени $t = 0$
Уравнение нормальной прямой к траектории движения тела в момент времени $t = 0$ можно записать, используя свойство перпендикулярности касательной и нормальной прямых. Если угловой коэффициент касательной $k$ определён как

$$k = \frac{s_{y\_derivative}}{s_{x\_derivative}},$$ то угловой коэффициент нормальной прямой $k_{normal}$ будет равен 

$$k_{normal} = -\frac{1}{k}$$

Таким образом, уравнение нормальной прямой, проходящей через точку $p_{start}(x_{start}, y_{start})$, имеет вид:

$$y - y_{start} = k_{normal}(x - x_{start})$$ или $$y = -\frac{1}{k}(x - x_{start}) + y_{start}$$

1. **Строфоида**: Полярное уравнение задано как $$\rho = \frac{a}{\cos(\phi)} + a \cdot \tan(\phi)$$
2. **Параметрические уравнения**: Заданы как $$x = \ln\left(\frac{1}{\sqrt{1 - t^4}}\right)$$ $$y = \arcsin\left(\frac{1 - t^2}{1 + t^2}\right)$$

```python
# Параметры для строфоиды
const_for_polar = 1 

# Полярное уравнение
phi = np.linspace(0, 2 * np.pi, 1000)
rho = (const_for_polar / np.cos(phi)) + const_for_polar * np.tan(phi)

# Убираем значения, которые приводят к бесконечности (косинус равен 0)
rho[np.abs(rho) > 10] = np.nan  # Заменяем большие значения для корректного отображения

# Преобразование в декартовы координаты
x_polar = rho * np.cos(phi)
y_polar = rho * np.sin(phi)

# Параметрические уравнения
t = np.linspace(-0.99, 0.99, 1000)  # Ограничиваем диапазон t, чтобы избежать деления на ноль
x_param = np.log(1 / np.sqrt(1 - t**4))
y_param = np.arcsin((1 - t**2) / (1 + t**2))

# Создание графиков
plt.figure()

# График строфоиды
plt.subplot(2, 1, 1, projection='polar')
plt.polar(phi, rho, color='green')
plt.title('Строфоида')

# График параметрических уравнений
plt.subplot(2, 1, 2)
plt.plot(x_param, y_param, color='blue')
plt.title('Параметрические уравнения')
plt.xlabel('x')
plt.ylabel('y')

plt.subplots_adjust(hspace=0.6)
plt.show()
```
